adding "Bold_6" from here: https://jiffyclub.github.io/palettable/ 

```python
import matplotlib.pyplot as plt
from palettable.cartocolors.qualitative import Bold_6 as mycm
plt.style.use("plotparam.mplstyle")
```

```python
europe = [1,2,3]
oceania = [3,6,8]
africa = [0,1,2]
america = [3,4,9]
australia = [2,4,8]
asia = [1,2,5]

plt.plot(europe, label = "europe")
plt.plot(oceania, label = "oceania")
plt.plot(africa, label = "africa")
plt.plot(america, label = "america")
plt.plot(asia, label = "asia")
plt.plot(australia, label = "australia")

plt.legend();
```

and then for options that are more result-specific we can have a dictionary like this


```python
plotparam = {
	"dpi": 96,
	"europe": {"color" : Bold_6.hex_colors[0], "linewidth" : 4, "label" : "EUR"},
    "oceania": {"color" : Bold_6.hex_colors[1], "s" : 20, "label": "OCE"},
	"africa": {"color" : Bold_6.hex_colors[2], "alpha": 0.3, "label": "AFR"},
	"asia": {"color" : Bold_6.hex_colors[3], "linewidth" : 4, "label" : "ASI"},
    "australia": {"color" : Bold_6.hex_colors[4], "s" : 20, "label": "AUS"},
	"america": {"color" : Bold_6.hex_colors[5], "alpha": 0.7, "label": "AME"}
			}
```

and then call the plotparam dict for plotting:

```python
# back to default settings (if needed)
plt.style.use("default")
```

```python
fig, ax = plt.subplots(1,1, figsize = (5,5))
ax.plot(europe, **plotparam["europe"])
ax.plot(america, **plotparam["america"])
ax.scatter([0,1,2], oceania, **plotparam["oceania"])
ax.scatter([2,3,4], australia, **plotparam["australia"])
ax.bar(x = africa, height = [4,5,7], **plotparam["africa"])
ax.bar(x = asia, height = [7,3,6], **plotparam["asia"])
ax.legend();
```
