here is a random example on the mplstyle stylesheet use. following what is explained here: https://matplotlib.org/stable/tutorials/introductory/customizing.html#customizing-with-matplotlibrc-files  

for the mplstyle sheet, i just took the default one (classic.mplstyle), commented it all out, and commented back in only the options that we want to edit

i guess this is for more "global" options like axes color and text font size etc. (example from michael: https://github.com/mszell/bikenwgrowth/blob/master/parameters/plotstyle.mplstyle )

```python
import matplotlib.pyplot as plt
plt.style.use("plotparam.mplstyle")
```

```python
europe = [1,2,3]
oceania = [3,6,8]
africa = [0,1,2]
plt.plot([1,2,3], label = "europe")
plt.plot([3,6,9], label = "oceania")
plt.plot([0,1,2], label = "africa")
plt.legend();
```

and then for options that are more result-specific we can have a dictionary like this


```python
plotparam = {
	"dpi": 96,
	"europe": {"color" : "grey", "linewidth" : 4, "label" : "EUR"},
    "oceania": {"color" : "black", "s" : 20, "label": "OCE"},
	"africa": {"color" : "green", "alpha": 0.3, "label": "AFR"}
			}
```

and then call the plotparam dict for plotting:

```python
# back to default settings (if needed)
plt.style.use("default")
```

```python
fig, ax = plt.subplots(1,1, figsize = (5,5))
ax.plot(europe, **plotparam["europe"])
ax.scatter([0,1,2], oceania, **plotparam["oceania"])
ax.bar(x = africa, height = [4,5,7], **plotparam["africa"])
ax.legend();
```
