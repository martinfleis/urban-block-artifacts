# Plotting styles specification

To ensure all the plots are following the same style, this notebook specifies it and shows examples.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Default styles based on seaborn themes

This uses seaborn to control the default style and a colormap.

The cell below is the only one that needs to be run. The rest should just work.

```python
import seaborn as sns
from palettable.cartocolors.qualitative import Bold_6


sns.set_theme(
    context="paper",
    style="ticks",
    rc={
        "patch.force_edgecolor": False,
        "axes.spines.top": False,
        "axes.spines.right": False,
        "axes.grid": True,
    },
    palette=Bold_6.hex_colors,
)
```

Plotting using the default will work in most cases.

```python
plt.plot(sorted(np.random.default_rng(42).gamma(np.linspace(0, 100, 100))))
plt.xlabel("meaningless index (order)")
plt.ylabel("something on Y axis")
plt.title("Cost of bananas")
```

```python
europe = [1, 2, 3]
oceania = [3, 6, 8]
africa = [0, 1, 2]
america = [3, 4, 9]
australia = [2, 4, 8]
asia = [1, 2, 5]

plt.plot(europe, label="europe")
plt.plot(oceania, label="oceania")
plt.plot(africa, label="africa")
plt.plot(america, label="america")
plt.plot(asia, label="asia")
plt.plot(australia, label="australia")

plt.legend()
```

```python
np.random.poisson(10, 100)
```

```python
s = pd.DataFrame(
    {
        "banana curviness": np.random.poisson(10, 100),
        "banana size": np.random.poisson(10, 100),
        "continent": (["Asia", "Africa", "South America", "Oceania", "Europe", "North America"] * 17)[:100]
    }
)
```

```python
sns.scatterplot(x="banana curviness", y="banana size", data=s, hue="continent");
```

If the grid is too noisy, you can manually override the settings.

```python
sns.scatterplot(x="banana curviness", y="banana size", data=s, hue="continent")
plt.grid(False)
```
